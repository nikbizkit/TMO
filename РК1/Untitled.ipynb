```python
import numpy as np
import pandas as pd
import os

DATA_PATH = os.path.join('fifa')
```

```python
def load_data(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, 'fifa.txt')
    return pd.read_csv(csv_path)
```

```python
data = load_data()
data
```

```python
PSO= data['PSO'].value_counts()
PSO
```

```python
sum_ = data['PSO'].count()
sum_
```

```python
data['PSO'].fillna('NO', inplace=True)
data
```

```python
mean = data['1st Goal'].mean()
data['1st Goal'].fillna(mean, inplace=True)
data
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(style='ticks')
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.scatterplot(ax=ax, x='Attempts', y='Blocked', data=data)
plt.title(r'Диаграмма рассеяния', fontsize=18, y=1.03);

plt.xlabel('удары', fontsize=14)
plt.ylabel('блоки', fontsize=14)
```
